```python
sns.histplot(x = "birth_rate", data = w_full, color = "violet", kde = True,  bins = 20 )
plt.title("Esperanza vida mundial")
plt.xlabel("observaciones")
plt.ylabel("Esperanza de vida")
```

La tasa de natalidad mas alta esta entre un 10 y 11, la mas baja entre un 42 y 45 

```python
sns.boxplot(x = "birth_rate", data = w_full, color = "violet")
plt.title("Tasa de natalidad mundial")
plt.ylabel("tasa de natalidad")
```

La mayoria se ubican entre un 11 y un 29
