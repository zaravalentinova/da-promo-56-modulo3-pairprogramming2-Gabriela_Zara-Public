Requisitos que deben cumplir las gráficas (en caso de usar alguna de ellas):

Todas deben llevar título.

Los nombres de los ejes deben estar en castellano.

En los scatterplot los puntos deben tener un tamaño de 2.

En los scatterplot los puntos deben tener un color diferente al azul.

En los barplot las barras siempre tienen que estar ordenadas de mayor a menor.

En los barplot y countplot las barras tienen que tener un color diferente a los que vienen por defecto.

En los pieplot siempre tenemos que añadir la etiqueta de a que % corresponde cada quesito.

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
```

```python
w_full = pd.read_csv("world_data_full_apply_nulos.csv", index_col=0)
```

```python
w_full.head()
```

1. ¿Cuál es el continente con la mayor cantidad de países en el conjunto de datos? Proporciona el porcentaje de países en ese continente.

```python
conteo_continente = w_full['continent'].value_counts()
```

```python
plt.figure(figsize=(8, 6))
conteo_continente.plot(
    kind='pie',
    autopct='%1.1f%%',     
    startangle=90,        
    colors=plt.cm.Paired.colors,  
    wedgeprops={'edgecolor': 'white'}
)
plt.title('Porcentaje de países por continente')
plt.ylabel('')  # Ocultar el eje Y
plt.tight_layout()
```

El continente con mas paises es Africa con un 27,3%

2. ¿Cuáles son los 10 países con la mayor área forestal?

```python
forestal = w_full.sort_values(by='forested_area', ascending=False).head(10)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=forestal, x='country', y='forested_area', palette='Greens_r')

plt.title('Top 10 países con mayor área forestal')
plt.xlabel('País')
plt.ylabel('Área forestal')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

3. ¿Cómo se distribuye la densidad de población entre los países?

4. ¿Existe una relación entre el PIB y la tasa de desempleo de los países?

```python
w_full['gdp_range'] = pd.qcut(w_full['gdp'], q=5, labels=['Muy Bajo', 'Bajo', 'Medio', 'Alto', 'Muy Alto'])

# Calcular promedio de tasa de desempleo para cada grupo de PIB
grouped = w_full.groupby('gdp_range')['unemployment_rate'].mean().reset_index()

# Graficar con barplot
plt.figure(figsize=(10,6))
sns.barplot(data=grouped, x='gdp_range', y='unemployment_rate', palette='Blues_d')

plt.title('Promedio de tasa de desempleo según rango de PIB')
plt.xlabel('Rango de PIB')
plt.ylabel('Tasa de desempleo promedio (%)')
plt.tight_layout()
plt.show()
```

5. ¿Cómo se relacionan la tasa de mortalidad infantil y la esperanza de vida?

```python
w_full['life_range'] = pd.cut(w_full['life_expectancy'], bins=[0, 50, 60, 70, 80, 90, 100],
                          labels=['0-50', '51-60', '61-70', '71-80', '81-90', '91-100'])
```

```python
grouped = w_full.groupby('life_range')['infant_mortality'].mean().reset_index()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=grouped, x='life_range', y='infant_mortality', palette='Reds_r')

plt.title('Promedio de mortalidad infantil por rango de esperanza de vida')
plt.xlabel('Rango de esperanza de vida (años)')
plt.ylabel('Mortalidad infantil promedio')
plt.tight_layout()
plt.show()
```
